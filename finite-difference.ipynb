# Second Order Derivitive for better approximation

```python
xmax = 10.0 #space
nx = 100 #discretize
a =0.25
dx = xspace/nx #increment
x = np.linspace(0,xspace,nx) #vector; iterates over x in plotting by xspace/nx increment
x0 = xmax/2

#define function
f = (1/np.sqrt(2*np.pi*a)) * np.exp(-((x-x0)**2)/(2*a)) 
plt.plot(x,f)
```

```python
nder3 = np.zeros(nx) # uses three points to estimate the second dervitive
ader = np.zeros(nx)

for i in range(1,nx-1):
    nder3[i] = (f[i+1] - 2*f[i] + f[i-1])/(dx**2)

ader = (1/np.sqrt(2*np.pi*a)) * ((x-x0)**2/a**2 - 1/a) * np.exp(-((x-x0)**2)/(2*a))

rms = np.sqrt(np.mean(nder3-ader)**2)

plt.plot(x,nder3)
plt.plot(x,ader)
plt.show()
```
