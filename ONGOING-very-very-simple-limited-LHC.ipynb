# 1. Particle Split

```python
def get_decay_angle(M, m1, m2, p1):
    # momentum conservation; initially M wasn't moving for simplicity
    # ignoring the weak force that resaults in y-axis split !
    # p2_x = p1_x
    # p2_y = -p1_y
    p2 = p1
    E1= (m1**2 * c**4 + p1**2 * c**2)**(1/2)
    E2= (m1**2 * c**4 + p1**2 * c**2)**(1/2)
    angle_cosined = (E1*E2 - (M**2 - m1**2 - m2**2)*((c**4) / 2)) / (p1*p2*c**2)
    decay_angle = np.arccos(angle_cosined)
    return decay_angle
```

```python
# constants
c=299792458
deg=180/np.pi

def decay(M, m1, m2, p1):
    x1_f, x2_f = 0,0 
    y1_f, y2_f = 0,0

    x1, y1 =[], []
    x2, y2 =[], []

    p2 = p1
    v1 = (2*p1 / m1) **(1/2)
    v2 = (2*p2 / m2) **(1/2)

    decay_angle = get_decay_angle(M, m1, m2, p1)

    # get velcities
    v1_x = v1 * np.cos(decay_angle/2)
    v1_y = v1 * np.sin(decay_angle/2)
    v2_x = v2 * np.cos(decay_angle/2)
    v2_y = - v2 * np.sin(decay_angle/2)

    dt = 0   
    run_time=10
    while dt < run_time:
        dt=dt*1
        
        
        if dt != 0:
            print(x1[-1] *2e5)
            print(x1[-1])        

        # get positions
        x1_f += v1_x * dt
        y1_f += v1_y * dt

        x2_f += v2_x * dt
        y2_f += v2_y * dt

        # save positions
        x1.append(x1_f)
        y1.append(y1_f)
        x2.append(x2_f)
        y2.append(y2_f)

        # pass time
        dt += 1
    x1 = np.array(x1)
    y1 = np.array(y1)
    x2 = np.array(x2)
    y2 = np.array(y2)
    return x1, y1, x2, y2
```
